```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/c/GeeksforGeeksVideos/videos')
driver.implicitly_wait(10)

```

```python
import pandas as pd
import numpy as np
```

```python
html_content = driver.page_source
soup = BeautifulSoup(html_content, 'html.parser')
video_titles = soup.find_all('yt-formatted-string', class_='style-scope ytd-rich-grid-media')
for title in video_titles:
    print(title.text)
```

```python
def extract_views(soup):
    try:
        views_element = soup.find_all('span', class_='style-scope ytd-grid-video-renderer')[0]
        return views_element.text.strip()
    except (AttributeError, IndexError):
        return np.nan
def extract_date(soup):
    try:
        date_element = soup.find_all('span', class_='inline-metadata-item style-scope ytd-video-meta-block')[1]
        return date_element.text.strip()
    except IndexError:
        return np.nan
```

```python
import numpy as np

data = []

for sp in soup.find_all('ytd-rich-item-renderer'):
    try:
        title_element = sp.find('yt-formatted-string', class_='style-scope ytd-rich-grid-media')
        video_title = title_element.text.strip() if title_element else np.nan
    except AttributeError:
        video_title = np.nan
    
    try:
        views_element = sp.find_all('span', class_='inline-metadata-item style-scope ytd-video-meta-block')[0]
        views = views_element.text.strip()
    except (AttributeError, IndexError):
        views = np.nan

    try:
        date_element = sp.find_all('span', class_='inline-metadata-item style-scope ytd-video-meta-block')[1]
        date_time = date_element.text.strip()
    except IndexError:
        date_time = np.nan

    try:
        video_link = sp.find('a', class_='yt-simple-endpoint')['href']
    except (KeyError, TypeError):
        video_link = np.nan

    try:
        thumbnail_link = sp.find('img', class_='style-scope yt-img-shadow')['src']
    except (KeyError, TypeError):
        thumbnail_link = np.nan

    try:
        img_link = sp.find('img', class_='yt-core-image--fill-parent-height')['src']
    except (KeyError, TypeError):
        img_link = np.nan

    data.append([video_title,views,date_time,video_link,img_link
                ])
```
